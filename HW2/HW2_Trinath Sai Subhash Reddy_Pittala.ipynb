## B) With no discount (gamma = 1), no living reward, and no noise, use the Value Iteration Algorithm with 100 iterations to update the optimal values for each state and print the result

```python
V = np.full((4, 5, 5), 0, dtype=float)
V[:,4,4] = 100
V[:,2,3] = -1000


# F forward one step, 2F forward 2 steps, ACW rotate anticlockwise, CW rotate Clockwise
action = {"F","2F","ACW","CW"}
actions = {"F": -1, "2F": -1.5, "ACW": -0.3, "CW": -0.3}

# Movement with black square and walls into consideration 
# 1: up, 2: down, 3: left, and 4: right
def movement(x, y, direction, action):
    if action == "F":
        if direction == 1 and x <= 4:
            if y + 1 <= 4:
                if not ((x == 1 and y + 1 == 4) or ((x == 0 and y == 1) and (x == 0 and y + 1 == 2)) or ((x == 4 and y == 1) and (x == 4 and y + 1 == 2))):
                    return x, y + 1, direction
        elif direction == 2 and x >= 0:
            if y - 1 >= 0:
                if not ((x == 1 and y - 1 == 4) or ((x == 0 and y == 2) and (x == 0 and y - 1 == 1)) or ((x == 4 and y == 2) and (x == 4 and y - 1 == 1))):
                    return x, y - 1, direction
        elif direction == 3 and y >= 0:
            if x - 1 >= 0:
                if not (x -1 == 1 and y == 4):
                    return x - 1, y, direction
        elif direction == 4 and y <= 4:
            if x + 1 <= 4:
                if not (x + 1 == 1 and y == 4):
                    return x + 1, y, direction
        else:
            return None

    elif action == "2F":
        if movement(x, y, direction, "F") is not None:
            x1, y1, d1 = movement(x, y, direction, "F")
            if (movement(x1, y1, d1, "F") is not None) and not (x1 == 2 and y1 == 3):
                x2, y2, d2 = movement(x1, y1, d1, "F")
                if not (x2 == 2 and y == 3):
                    return x2, y2, d2
        else:
            return None

        
    elif action == "ACW":
        if direction == 1:
            return x, y, 3
        elif direction == 2:
            return x, y, 4
        elif direction == 3:
            return x, y, 2
        elif direction == 4:
            return x, y, 1    
    
    elif action == "CW":
        if direction == 1:
            return x, y, 4
        elif direction == 2:
            return x, y, 3
        elif direction == 3:
            return x, y, 1
        elif direction == 4:
            return x, y, 2
    
    else:
        return None
```

```python
def value(x, y, direction, gamma):
    action_values = []
    action_list = []

    for action, cost in actions.items():
        new_pos = movement(x, y, direction, action)
        if new_pos is not None:
            x_new, y_new, direction_new = new_pos
            expected_value = cost + (gamma * V[direction_new - 1, x_new, y_new])
            action_values.append(expected_value)
            action_list.append(action)

    max_index = np.argmax(action_values)
    state_value = np.max(action_values)
    best_action = max_index + 1

    return state_value, best_action
```

```python
def valueIterationPrint(V,best_actions):
    for direction in range(4):
            for x in range(5):
                for y in range(5):
                    n = V[direction][x][y]
                    s = f"{n:.9f}"
                    if best_actions[direction][x][y] == 1:
                        print("State(" + str(x + 1) + "," + str(y + 1) + "," + str(direction + 1) + ")" + "  V =  " + str(s) + " " + "Best Action " + "A1")
                    if best_actions[direction][x][y] == 2:
                        print("State(" + str(x + 1) + "," + str(y + 1) + "," + str(direction + 1) + ")" + "  V =  " + str(s) + " " + "Best Action " + "A2")
                    if best_actions[direction][x][y] == 3:
                        print("State(" + str(x + 1) + "," + str(y + 1) + "," + str(direction + 1) + ")" + "  V =  " + str(s) + " " + "Best Action " + "A3")
                    if best_actions[direction][x][y] == 4:
                        print("State(" + str(x + 1) + "," + str(y + 1) + "," + str(direction + 1) + ")" + "  V =  " + str(s) + " " + "Best Action " + "A4")
                    if best_actions[direction][x][y] == -1:
                        print("State(" + str(x + 1) + "," + str(y + 1) + "," + str(direction + 1) + ")" + "  V =  " + str(s) + " " + "Terminal or Block State")
```

```python
gamma = 1
max_iterations = 100
theta = 0.0001
deltathetabool = False
for i in range(max_iterations):

    delta = 0

    Vnew = V.copy()
    best_actions_new = np.full((4, 5, 5), -1, dtype=int)
    for direction in range(1, 5):
        for x in range(5):
            for y in range(5):
                if (x, y) == (4, 4) or (x, y) == (2, 3) or (x, y) == (1,4):
                    continue

                new_value, best_action = value(x, y, direction, gamma)

                delta = max(delta, abs(new_value - Vnew[direction - 1, x, y]))

                Vnew[direction - 1, x, y] = new_value
                best_actions_new[direction - 1, x, y] = best_action
    
    V = Vnew.copy()
    best_actions = best_actions_new.copy()
    if i<3:
        print("Iteration "+str(i+1)+":")
        i = i + 1
        valueIterationPrint(Vnew,best_actions_new)
        print("--------------------------------------")
    if not deltathetabool:
        if delta < theta:
            print("No of iteration before delta < theta: " + str(i))
            deltathetabool = True

print("Final Iteration after 100 times "+":")        
valueIterationPrint(Vnew,best_actions_new)
```

## F) Repeat part B, but this time with noise = 0.1, and gamma = 0.9 and no living reward. With a noise of 0.1, every time you take an action, the result will be the expected action with probability 0.9 (90%), but 10% of the time, the robot will instead take a different action 

```python
def value_noise(x, y, direction, gamma, noise):
    action_values = []
    action_list = []

    for action, cost in actions.items():
        new_pos = movement(x, y, direction, action)
        
        if new_pos is not None:
            x_new, y_new, direction_new = new_pos
            expected_value = cost + (1 - noise) * (gamma * V[direction_new - 1, x_new, y_new])
            if noise > 0:
                unexpected_actions = [a for a in actions.keys() if a not in action_list]
                num_unexpected_actions = len(unexpected_actions)
                if num_unexpected_actions > 0:
                    unexpected_value = 0
                    for a in unexpected_actions:
                        new_pos = movement(x, y, direction, a)
                        if new_pos is not None:
                            x_new, y_new, direction_new = new_pos
                            unexpected_value += gamma * V[direction_new - 1, x_new, y_new]
                    unexpected_value /= num_unexpected_actions
                    expected_value += noise * unexpected_value

            action_values.append(expected_value)
            action_list.append(action)

    max_index = np.argmax(action_values)
    state_value = np.max(action_values)
    best_action = max_index + 1

    return state_value, best_action
```

```python
V = np.full((4, 5, 5), 0, dtype=float)
V[:,4,4] = 100
V[:,2,3] = -1000

gamma = 0.9
noise = 0.1
max_iterations = 100
theta = 0.0001
deltathetabool = False
for i in range(max_iterations):

    delta = 0

    Vnew = V.copy()
    best_actions_new = np.full((4, 5, 5), -1, dtype=int)
    for direction in range(1, 5):
        for x in range(5):
            for y in range(5):
                if (x, y) == (4, 4) or (x, y) == (2, 3) or (x, y) == (1,4):
                    continue

                new_value, best_action = value_noise(x, y, direction, gamma, noise)

                delta = max(delta, abs(new_value - Vnew[direction - 1, x, y]))

                Vnew[direction - 1, x, y] = new_value
                best_actions_new[direction - 1, x, y] = best_action
    
    V = Vnew.copy()
    best_actions = best_actions_new.copy()
    if i<3:
        print("Iteration "+str(i+1)+":")
        i = i + 1
        valueIterationPrint(Vnew,best_actions_new)
        print("--------------------------------------")
    if not deltathetabool:
        if delta < theta:
            print("No of iteration before delta < theta: " + str(i))
            deltathetabool = True


print("Final Iteration after 100 times "+":")        
valueIterationPrint(Vnew,best_actions_new)

```
